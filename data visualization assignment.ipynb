```python
import seaborn as sns
```

```python
a=sns.load_dataset("iris")
```

```python
a
```

```python
sns.scatterplot(x = "sepal_width",y="petal_length",data = a)#scatter plot is used to show the distribution this plot shows that there are 2 groups of flowers that have greater sepal width have lesser petal length and vice versa
```

```python
sns.lineplot(x='petal_length',y='sepal_length',data=a)#lineplot shows trends and we notice that as the petal length increases the sepal length also increases
```

```python
sns.distplot(a['petal_length'])#distribution plot here we can see that the highest no of flowers have a petal length 5 ,there are only a few flowers with petal length of 3 and a larger number of flowers with petal length of 1 
```

```python
sns.distplot(a['sepal_length'])#here we can see the highest number of flowers have a sepal length between 5 to 7 
```

```python
sns.distplot(a['sepal_width'])#here we can see the highest number of flowers have a sepal width of 3.0 
```

```python
sns.distplot(a['petal_width'])# here we can see that the highest no of flowers have a petal width of 1.5 ,there are only a few flowers with petal width between 0.5 to 1.0 and a larger number of flowers with petal width of 0.0 to 0.5, 
```

```python
sns.barplot(x= "species",y = "sepal_width",data = a)#here we can see that for the setosa species the sepal width is the highest
```

```python
sns.countplot(x= "species",data = a)#there are 50 species of each catogory 
```

```python
sns.boxplot(x = "species",y = "sepal_length",data = a)#there is only one outlier in the virginica family 
```
