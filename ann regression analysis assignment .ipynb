```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer([('one',OneHotEncoder(),[3])],remainder='passthrough')
```

```python
x=ct.fit_transform(x)
```

```python
x.shape
```

```python
x[0]
```

```python
#split data into x and y 
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
print(x.shape,"xshape")
print(y.shape,"yshape")
print(x_train.shape,"x_train shape")
print(x_test.shape,"x_test shape")
print(y_train.shape,"y_train shape")
print(y_test.shape,"y_test shape")
```

```python
from sklearn.preprocessing import StandardScaler
d=StandardScaler()
x_train=d.fit_transform(x_train)
x_test=d.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#model building 
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense 
```

```python
rg= Sequential()
```

```python
x_train.shape
```

```python
rg.add(Dense(units= 6,kernel_initializer='random_uniform',activation='relu'))
```

```python
rg.add(Dense(units=12,kernel_initializer='random_uniform',activation='relu'))
```

```python
rg.add(Dense(units= 1,kernel_initializer='random_uniform',activation='relu'))
```

```python
rg.compile(optimizer='adam',loss='mse', metrics=['mse'])
```

```python
rg.fit(x_train,y_train,epochs=10000)#model evaluation
```

```python
ypred = rg.predict(x_test)
```
